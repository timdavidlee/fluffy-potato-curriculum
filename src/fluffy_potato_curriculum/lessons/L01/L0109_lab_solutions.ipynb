## 1. Setup

One live client, used for every problem below. With a real model you'll see genuine sampling spread — it won't be identical run to run, which is the whole point.

````python
from fluffy_potato_curriculum.potato_llm import AnthropicClient, Message, PotatoLLMClient

# Live client. The key is read through fluffy_potato_curriculum.common.config
# (a pydantic-settings layer over the environment / .env); constructing the client
# raises a clear error if ANTHROPIC_API_KEY is missing, rather than failing mid-call.
client: PotatoLLMClient = AnthropicClient()

PROMPT = "Suggest a name for a coffee shop on a rainy street in Seattle. Just the name, one or two words."
print(f"setup ready (live client: {client.model})")
````

## 2. Problem 1 — Sample at temperature 0

Write `sample_n(client, temperature, n)`: call `client.chat([Message.user(PROMPT)], max_tokens=16, temperature=temperature)` n times and collect the replies. Run 5x at 0.

````python
def sample_n(client: PotatoLLMClient, temperature: float, n: int) -> list[str]:
    return [
        client.chat([Message.user(PROMPT)], max_tokens=16, temperature=temperature).text.strip()
        for _ in range(n)
    ]


for line in sample_n(client, 0.0, 5):
    print(line)
````

## 3. Problem 2 — Sample at temperature 1

Run the same prompt 5x at temperature 1. How does the spread compare?

````python
for line in sample_n(client, 1.0, 5):
    print(line)
````
