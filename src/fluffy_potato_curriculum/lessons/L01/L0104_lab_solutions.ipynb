## Setup

Runs offline (tiktoken downloads its vocabulary once on first use).

````python
import tiktoken

SAMPLES = {
    "english": "The quick brown fox jumps over the lazy dog.",
    "code": "def f(x): return x ** 2",
    "json": '{"user_id": 12345, "events": ["click", "scroll"]}',
}
print("setup ready")
````

### Problem 1 — Count tokens

Implement `count_tokens` using tiktoken's `cl100k_base` encoding, then print the count for each
sample. Before you run it: which sample do you *guess* is longest in tokens?

````python
def count_tokens(text: str, encoding_name: str = "cl100k_base") -> int:
    enc = tiktoken.get_encoding(encoding_name)
    return len(enc.encode(text))


for name, text in SAMPLES.items():
    print(name, "->", count_tokens(text))
````

### Problem 2 — See the boundaries

Implement `show_boundaries`: decode each token on its own and join the pieces with `|`. This is the
view that makes "a token is not a word" obvious.

````python
def show_boundaries(text: str, encoding_name: str = "cl100k_base") -> str:
    enc = tiktoken.get_encoding(encoding_name)
    return "|".join(enc.decode([token_id]) for token_id in enc.encode(text))


print(show_boundaries(SAMPLES["code"]))
````

### Problem 3 — Tokenizers disagree

Compute the token count of each sample under two encodings — `cl100k_base` (GPT-3.5/4) and
`o200k_base` (GPT-4o) — and print them side by side. Then answer in the markdown cell below.

````python
for name, text in SAMPLES.items():
    cl = count_tokens(text, "cl100k_base")
    o2 = count_tokens(text, "o200k_base")
    print(f"{name:9} cl100k={cl:3}  o200k={o2:3}")
````

### Problem 4 — Where the rule of thumb breaks

Compute the characters-per-token ratio for the `english` and `json` samples. The rule of thumb says
≈ 4.0. State which sample breaks it and why.

````python
for name in ("english", "json"):
    text = SAMPLES[name]
    ratio = len(text) / count_tokens(text)
    print(f"{name:9} {ratio:.2f} chars/token")
# english is ~4 chars/token; json is far denser because every brace, quote, colon,
# and comma becomes its own token, so its chars/token ratio is well below 4.
````
