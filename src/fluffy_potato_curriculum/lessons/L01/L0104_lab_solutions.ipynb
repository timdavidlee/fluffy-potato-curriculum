## 1. Setup

## 2. Problem 1 — Count tokens

Implement `count_tokens` with tiktoken's `cl100k_base`. Guess which sample is longest *before* running.

````python
def count_tokens(text: str, encoding_name: str = "cl100k_base") -> int:
    enc = tiktoken.get_encoding(encoding_name)
    return len(enc.encode(text))


for name, text in SAMPLES.items():
    print(name, "->", count_tokens(text))
````

## 3. Problem 2 — See the boundaries

Implement `show_boundaries`: decode each token on its own, join with `|`.

````python
def show_boundaries(text: str, encoding_name: str = "cl100k_base") -> str:
    enc = tiktoken.get_encoding(encoding_name)
    return "|".join(enc.decode([token_id]) for token_id in enc.encode(text))


print(show_boundaries(SAMPLES["code"]))
````

## 4. Problem 3 — Tokenizers disagree

Count each sample under `cl100k_base` and `o200k_base`. Then answer below: why does "how many tokens is this?" have no single universal answer?

````python
for name, text in SAMPLES.items():
    cl = count_tokens(text, "cl100k_base")
    o2 = count_tokens(text, "o200k_base")
    print(f"{name:9} cl100k={cl:3}  o200k={o2:3}")
````

## 5. Problem 4 — Where the rule of thumb breaks

Compute chars-per-token for `english` and `json`. Which breaks the ≈4.0 rule, and why?

````python
for name in ("english", "json"):
    text = SAMPLES[name]
    ratio = len(text) / count_tokens(text)
    print(f"{name:9} {ratio:.2f} chars/token")
# json is far denser: every brace, quote, colon, and comma is its own token, so its
# chars/token ratio sits well below the ~4.0 that English prose follows.
````
