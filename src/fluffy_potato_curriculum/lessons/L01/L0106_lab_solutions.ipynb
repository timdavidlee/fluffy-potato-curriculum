## Setup

````python
import os

from fluffy_potato_curriculum.potato_llm import AnthropicClient, Message

HAS_KEY: bool = bool(os.environ.get("ANTHROPIC_API_KEY"))
PROMPT = "Suggest a name for a coffee shop on a rainy street in Seattle. Just the name, one or two words."
print("ANTHROPIC_API_KEY set:", HAS_KEY)
````

### Problem 1 — Five runs at temperature 0

Write `sample_n(temperature, n)` that sends `PROMPT` `n` times at the given temperature (use
`max_tokens=16`) and returns the list of replies. Run it 5 times at `temperature=0` and look at the
spread.

````python
def sample_n(temperature: float, n: int) -> list[str]:
    if not HAS_KEY:
        return [f"<set ANTHROPIC_API_KEY> temp={temperature}"] * n
    client = AnthropicClient()
    replies: list[str] = []
    for _ in range(n):
        reply = client.chat([Message.user(PROMPT)], max_tokens=16, temperature=temperature)
        replies.append(reply.text.strip())
    return replies


for line in sample_n(0.0, 5):
    print(line)
````

### Problem 2 — Five runs at temperature 1

Run the same prompt 5 times at `temperature=1.0`. How does the spread compare to Problem 1?

````python
for line in sample_n(1.0, 5):
    print(line)
````
