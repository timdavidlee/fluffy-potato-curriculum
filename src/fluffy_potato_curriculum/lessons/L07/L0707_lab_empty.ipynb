````python
from langchain_core.messages import AIMessage, BaseMessage, HumanMessage, ToolMessage
from langchain_core.messages.tool import ToolCall

CALL_ID = "toolu_01XYZ"
transcript: list[BaseMessage] = [
    HumanMessage(content="What is 6,150 multiplied by 7,012?"),
    AIMessage(
        content="",
        tool_calls=[
            ToolCall(
                id=CALL_ID, name="calculator", args={"expression": "6150 * 7012"}, type="tool_call"
            )
        ],
    ),
    ToolMessage(content="43123800", tool_call_id=CALL_ID),
    AIMessage(content="6,150 x 7,012 = 43,123,800."),
]
print(f"{len(transcript)} messages captured")
````
