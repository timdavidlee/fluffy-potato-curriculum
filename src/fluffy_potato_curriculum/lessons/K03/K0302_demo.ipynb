A tiny notebook to practice the Jupyter workflow. Do **Kernel → Restart Kernel and Run All
Cells** and watch every cell run top-to-bottom with execution counts `1, 2, 3, …` and no
errors. That clean top-to-bottom run is the habit K03 is about.

````python
# The first code cell is the setup cell: all imports live here.
import asyncio
import math
````

````python
# A normal computation that uses the setup cell above it.
radius = 3
area = math.pi * radius**2
print(f"circle area = {area:.2f}")
````

````python
# You can `await` directly in a Jupyter cell — no asyncio.run(...) needed.
# K05 leans on this; here it's just proof the kernel supports it.
await asyncio.sleep(0.05)
print("awaited directly in a cell — Jupyter supports top-level await")
````

## Done

If every cell ran in order with no errors, you've got the workflow. Whenever a notebook acts
weird, **Restart Kernel and Run All Cells** is your reset button — it proves the notebook
works from a clean kernel, not just in the scrambled order you happened to run cells.
