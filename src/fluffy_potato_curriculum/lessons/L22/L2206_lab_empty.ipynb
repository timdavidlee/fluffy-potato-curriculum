## 1. Setup

This capstone is mostly **judgment**, with one live look at a real skill. The setup is **given**: the decision heuristic (for reference), a helper that splits a `SKILL.md` into its frontmatter and body, and a path to the repo root. Run it as-is.

````python
from __future__ import annotations

from pathlib import Path

import fluffy_potato_curriculum

# The repo root, resolved from the installed package (works from any cwd).
REPO_ROOT = Path(fluffy_potato_curriculum.__file__).resolve().parents[2]

# The decision heuristic, for reference while you classify.
HEURISTIC = {
    "tool": "deterministic operation, structured input/output, called mid-reasoning",
    "system prompt": "always-true behavior / persona / a rule that must hold on EVERY call",
    "skill": "multi-step procedure or know-how used only SOMETIMES, maybe bundling scripts",
}


def parse_skill_md(path: Path) -> tuple[dict[str, str], str]:
    """Split a SKILL.md into (frontmatter dict, body).

    Frontmatter is the block between the first two `---` lines; the rest is the body.
    Example return:
        ({"name": "author-lesson-roadmap", "description": "Draft ..."}, "# Author ...")
    """
    text = path.read_text(encoding="utf-8")
    _, front_text, body = text.split("---", 2)
    front: dict[str, str] = {}
    for line in front_text.strip().splitlines():
        key, _, value = line.partition(":")
        front[key.strip()] = value.strip()
    return front, body.strip()


for home, rule in HEURISTIC.items():
    print(f"{home:>14}: {rule}")
````
