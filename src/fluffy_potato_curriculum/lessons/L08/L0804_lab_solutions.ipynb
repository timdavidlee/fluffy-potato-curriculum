## 1. Setup — the signals and a table of tasks

The four **'tool warranted'** signals and two **'no tool'** signals from the lecture, plus a table of candidate tasks. No imports needed — this is reasoning in Python data structures.

````python
# The four 'reach for a tool' signals (objectives.md), as short tags:
WARRANTED = {
    "cant_memorize": "data the model can't have memorized (live state, recent events)",
    "bad_at_compute": "precise computation the model is bad at (big arithmetic, date math)",
    "side_effect": "a side effect outside the conversation (email, DB write, external API)",
    "verify": "verification against ground truth (database lookup, fact retrieval)",
}
# The two 'no tool' signals:
NO_TOOL = {
    "model_knows": "the model already knows it with high reliability",
    "not_worth_roundtrip": "the round-trip costs more than the marginal accuracy gain",
}

TASKS = {
    "T1": "What is the capital of France?",
    "T2": "What is the current temperature in Seattle right now?",
    "T3": "How many days are between 1987-03-12 and 2024-11-04?",
    "T4": "Send a Slack message to the on-call engineer.",
    "T5": "Does user u_8841 have an active subscription?",
    "T6": "Summarize the three sentences I just pasted.",
}
print(len(TASKS), "tasks;", len(WARRANTED), "warranted signals;", len(NO_TOOL), "no-tool signals")
````

## 2. Problem 1 — Decide tool-or-not for each task

Fill in `DECISIONS`: for each task id, a dict with `tool` (`True`/`False`) and `signal` (the matching key from `WARRANTED` if a tool, or from `NO_TOOL` if not). One signal each — pick the *dominant* one.

Hint: T1 the model owns cold; T2 needs live data; T3 is exact date math; T4 has a side effect; T5 verifies against a database; T6 is something the model does well on its own.

````python
DECISIONS: dict[str, dict[str, object]] = {
    "T1": {"tool": False, "signal": "model_knows"},
    "T2": {"tool": True, "signal": "cant_memorize"},
    "T3": {"tool": True, "signal": "bad_at_compute"},
    "T4": {"tool": True, "signal": "side_effect"},
    "T5": {"tool": True, "signal": "verify"},
    "T6": {"tool": False, "signal": "not_worth_roundtrip"},
}
````

## 3. Problem 2 — Check your signals are valid

Implement `valid_signal(decision)`: a decision is internally consistent when a `tool: True` uses a key from `WARRANTED`, and a `tool: False` uses a key from `NO_TOOL`. Run it over every decision — all should be `True`.

````python
def valid_signal(decision: dict[str, object]) -> bool:
    """True iff a tool decision uses a WARRANTED signal, a no-tool a NO_TOOL signal."""
    if decision["tool"]:
        return decision["signal"] in WARRANTED
    return decision["signal"] in NO_TOOL


for tid, d in DECISIONS.items():
    print(tid, "->", valid_signal(d))
````

## 4. Problem 3 — Defend each call in one sentence

Implement `defend(task_id)`: return one sentence of the form `"Tool: <why>."` or `"No tool: <why>."`, using the signal's description text. Naming *why* is the whole skill. Print the defense for every task.

````python
def defend(task_id: str) -> str:
    """One-sentence justification using the chosen signal's description."""
    d = DECISIONS[task_id]
    if d["tool"]:
        return f"Tool: {WARRANTED[d['signal']]}."
    return f"No tool: {NO_TOOL[d['signal']]}."


for tid in TASKS:
    print(f"{tid}: {defend(tid)}")
````
