## 1. Setup — a registry of tools to assess

Six tools, each with a one-line behavior note. Some are read-only or pure; some have a side effect outside the conversation. No imports needed.

````python
TOOLS = {
    "get_weather": "read-only lookup of current weather",
    "add_numbers": "pure computation, no I/O",
    "charge_card": "charges the user's card for an amount",
    "send_email": "sends an email to a recipient",
    "upsert_contact": "writes a contact keyed by a stable contact_id",
    "create_ticket": "creates a new support ticket (new id each call)",
}
print(len(TOOLS), "tools to assess")
````

## 2. Problem 1 — Tag each tool safe-or-unsafe to retry

Fill in `SAFE_TO_RETRY`: a dict from tool name to `True` (safe to call twice — read-only, pure, or idempotent-with-a-key) or `False` (a second call double-acts). Recall the lecture's table: read-only / pure / keyed-upsert are safe; sending / charging / create-without-key are not.

````python
SAFE_TO_RETRY: dict[str, bool] = {
    "get_weather": True,  # read-only
    "add_numbers": True,  # pure
    "charge_card": False,  # double charge
    "send_email": False,  # duplicate email
    "upsert_contact": True,  # stable key -> idempotent
    "create_ticket": False,  # a second ticket each call
}
````

## 3. Problem 2 — Recommend a mitigation for each unsafe tool

Implement `mitigation(tool)` for the *unsafe* tools, returning one of the three lecture mitigations: `"idempotency_key"`, `"confirmation_step"`, or `"warn_in_description"`. Any defensible choice is fine — e.g. an idempotency key for `create_ticket`, a confirmation step for `charge_card`. Safe tools return `"none_needed"`.

````python
MITIGATIONS = {
    "charge_card": "confirmation_step",
    "send_email": "idempotency_key",
    "create_ticket": "idempotency_key",
}


def mitigation(tool: str) -> str:
    """A defensible mitigation per unsafe tool; safe tools need none."""
    if SAFE_TO_RETRY[tool]:
        return "none_needed"
    return MITIGATIONS[tool]


for t in TOOLS:
    print(f"{t:16} safe={SAFE_TO_RETRY[t]!s:5} -> {mitigation(t)}")
````

## 4. Problem 3 — Show a duplicate slips through without a key

Implement `send_email(to, body, sent)` where `sent` is a list acting as a sent-log: each call appends `(to, body)` and returns the new log length. Call it **twice** with identical arguments and show the log has **two** entries — the non-idempotent failure the lecture's OUTBOX demo showed live.

````python
def send_email(to: str, body: str, sent: list[tuple[str, str]]) -> int:
    """Non-idempotent send: each call appends another entry to the sent-log."""
    sent.append((to, body))
    return len(sent)


log: list[tuple[str, str]] = []
send_email("a@b.com", "hi", log)
send_email("a@b.com", "hi", log)  # model retried 'just in case'
print("log length:", len(log), "(expect 2 — duplicate sent)")
````
